```julia
using CSV, DataFrames

x = CSV.read("D:\\Dropbox (UCL)\\Rotation_vte\\Locomotion\\analysis\\ephys\\GLM\\unit_100\\FC_220410_BAA1101192_hairpin_unit_100.csv", DataFrame)
```

```julia
using Parquet
path = "D:\\Dropbox (UCL)\\Rotation_vte\\Locomotion\\analysis\\ephys\\GLM\\unit_100\\FC_220410_BAA1101192_hairpin_unit_100.parquet"
df = DataFrame(read_parquet(path))
```

```julia
names(df)
```

```julia
using  GLM


# rename!(df,Symbol("p(spike|ms)") => :y)
formula_complete = @formula(
    p ~ s + v + v^2 + dv_250ms + dv_500ms + dv_1000ms + omega + omega^2 + domega_250ms + domega_500ms + domega_1000ms + curv_0cm + curv_10cm + curv_20cm + curv_30cm
)

```

```julia

model = glm(formula_complete, df, Binomial(), LogitLink())
ytrue = x[:, :y]
yhat = predict(model, x);
```
